```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.style.use('classic')
#mpl.use('pdf')
mpl.rcParams['xtick.major.size'] = 8
mpl.rcParams['xtick.major.width'] = 2
mpl.rcParams['ytick.major.size'] = 8
mpl.rcParams['ytick.major.width'] = 2
mpl.rcParams['axes.linewidth']= 2
file=['sasa']#,'hbonds_pnw','hbonds_pnp','radius_of_gyration','rmsd','sasa']
for a in file:
    data1=np.genfromtxt('../wild/simulation_results/'+a+'.dat')
    data2=np.genfromtxt('../positive_charge36/simulation_results/'+a+'.dat')
    data3=np.genfromtxt('../negative_charge36/simulation_results/'+a+'.dat')
    plt.figure(facecolor='white')
    plt.title(a,weight='bold')
    plt.plot(data1[:,0], data1[:,1], color='green', label="Wild type")
    plt.plot(data2[:,0], data2[:,1], color='blue', label="Positive charged")
    plt.plot(data3[:,0], data3[:,1], color='red', label="Negative Charged")
    plt.xticks(rotation=75,weight='bold')
    #plt.yticks(np.arange(0,11,1),weight='bold')
    plt.xlabel("Frame Number",weight='bold')
    plt.ylabel("SASA (^2)",weight='bold') #"RGYR()"
    plt.legend(loc='best')
    plotfile=a+'.png'
    plt.savefig(plotfile, dpi=300, bbox_inches='tight')
```

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
plt.style.use('classic')
#mpl.use('pdf')
mpl.rcParams['xtick.major.size'] = 8
mpl.rcParams['xtick.major.width'] = 2
mpl.rcParams['ytick.major.size'] = 8
mpl.rcParams['ytick.major.width'] = 2
mpl.rcParams['axes.linewidth']= 2
file=['radius_of_gyration']#,'hbonds_pnw','hbonds_pnp','radius_of_gyration','rmsd','sasa']
for a in file:
    data1=np.genfromtxt('../wild/simulation_results/radius_of_gyration.dat')
    data2=np.genfromtxt('../positive_charge36/simulation_results/radius_of_gyration.dat')
    data3=np.genfromtxt('../negative_charge36/simulation_results/radius_of_gyration.dat')
    numbers1=np.arange(0,(max(data1[:,0])+1)/10,0.1)
    numbers2=np.arange(0,(max(data1[:,0])+1)/10,0.1)
    numbers3=np.arange(0,(max(data1[:,0])+1)/10,0.1)
    plt.figure(facecolor='white', figsize=(15,10))
    #plt.title("Radius of Gyration",fontsize=18,weight='bold')
    plt.plot(numbers1, data1[:,1], color='green', label="WT-GFP", lw=2)
    plt.plot(numbers2, data2[:,1], color='blue', label="SP-GFP",lw=2)
    plt.plot(numbers3, data3[:,1], color='red', label="SN-GFP",lw=2)
    plt.xticks(np.arange(0,220,20),rotation=0,fontsize=18,weight='bold')
    plt.yticks(np.arange(0,32,2),fontsize=18,weight='bold')
    plt.xlabel("Time (ns)",fontsize=18,weight='bold')
    plt.ylabel(r"R$_g$ ()",fontsize=18,weight='bold') #"RGYR()"
    plt.legend(numpoints = 1,fontsize=98, prop=dict(weight='bold',size=18),loc='best')
    # Set the minor ticks
    ax = plt.gca()
    x_minor_locator = plt.MultipleLocator(5)
    y_minor_locator = plt.MultipleLocator(0.5)
    ax.xaxis.set_minor_locator(x_minor_locator)
    ax.yaxis.set_minor_locator(y_minor_locator)
    ax.tick_params(which='both', width=2)
    ax.tick_params(which='major', length=10)
    ax.tick_params(which='minor', length=5, color='black')
plotfile='radius_of_gyration-Bulk.png'
plt.savefig(plotfile, dpi=300, bbox_inches='tight')
plt.show()
```
